# Prepare Dinosaur data

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy
import scipy
import utils
```

## Load data into Python and save it

```python
mat = scipy.io.loadmat("dinosaur/dino_Ps.mat")
cameras = numpy.stack(mat["P"][0])

measurements = numpy.loadtxt("dinosaur/viff.xy")
measurements = measurements.reshape(-1, cameras.shape[0], 2)

plt.close("all")
fig = plt.figure(figsize=(6, 6))
ax = plt.axes(projection="3d")
for i, cam in enumerate(cameras):
    utils.show_cam(cam, ax, scale=0.3)
    
ax.view_init(elev=-166, azim=44)
ax.set_aspect("equal")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("Dinosaur")

plt.show()

utils.save_cameras("dinosaur/cameras.bin", cameras)
utils.save_measurements("dinosaur/measurements.bin", measurements)
```
